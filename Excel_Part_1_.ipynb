**Question 1**: What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly
explain with examples.

**Answer:**
Paste simply inserts copied data exactly as it is, including values, formulas, and formatting.
Paste Special allows users to paste specific components such as values only, formulas only, formats, or perform operations like transpose.

**Example:**

Paste: Copies formula =A1+B1 along with formatting

Paste Special → Values: Pastes only the result (e.g., 50) without the formula

**Question 2:** Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes'
in Excel.
**Answer:**

Freeze Panes locks selected rows or columns so they remain visible while scrolling.

Split Panes divides the worksheet into multiple sections that scroll independently.

**Question 3:** Explain the difference between inserting a new row and inserting a new
column in Excel. Can you insert multiple rows or columns at once?
**Answer:**

Inserting a row adds a new horizontal line of cells.

Inserting a column adds a new vertical line of cells.

Yes, multiple rows or columns can be inserted at once by selecting multiple rows/columns before clicking Insert.


**Question 4:** What are logical functions in Excel? Provide examples of at least two
logical functions and their applications.

**Answer:**
Logical functions perform tests and return TRUE or FALSE.

Examples:

IF(A1>50,"Pass","Fail") – Used for result evaluation

AND(A1>50,B1>50) – Used when multiple conditions must be satisfied

**Question 5:** Discuss the purpose of 'XLOOKUP' and how it differs from the traditional
'VLOOKUP' function

**Answer:**
XLOOKUP is a modern lookup function that replaces VLOOKUP.

Differences:

XLOOKUP works both left and right

Does not require column index number

Handles missing values better
